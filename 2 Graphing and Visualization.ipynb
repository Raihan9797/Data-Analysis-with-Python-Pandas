# 3.2 creating a list of MAs for each region. Index is time and the columns are each regions MAs!

```python
graph_df = pd.DataFrame()

for region in df['region'].unique()[:16]:
    print(region)
    region_df = df.copy()[df['region'] == region]
    region_df["Date"] = pd.to_datetime(region_df["Date"])
    region_df.set_index("Date", inplace = True)
    region_df.sort_index(inplace = True)
    region_df[region + "price25ma"] = region_df['AveragePrice'].rolling(25).mean()
    
    if graph_df.empty:
        graph_df = region_df[[region + "price25ma"]]
        
    else :
        graph_df = graph_df.join(region_df[region + "price25ma"])
```

## also when you run the function, see how it slows down near the end. Sentdex's explanation is that because of the "Type" of avocados ie there are "organic" and "conventional", this means that there are dates which are the same for different avo types. 

## this means when u join(), it gets harder and harder for pandas to find the correct points to join the tables!!

## to resolve this, let's just focus on organic avocadoes! ie redo # 3!!

```python
# redo 3

# only organic avocadoes!!!
df = df.copy()[df["type"] == "organic"]

df['Date'] = pd.to_datetime(df["Date"])
# sorting by cols instead of by index (we have not set the index yet!)
df.sort_values(by = "Date", ascending = True, inplace = True)

graph_df = pd.DataFrame()

for region in df['region'].unique():
    print(region)
    region_df = df.copy()[df['region'] == region]
    region_df["Date"] = pd.to_datetime(region_df["Date"])
    region_df.set_index("Date", inplace = True)
    region_df.sort_index(inplace = True)
    region_df[region + "price25ma"] = region_df['AveragePrice'].rolling(25).mean()
    
    if graph_df.empty:
        graph_df = region_df[[region + "price25ma"]]
        
    else :
        graph_df = graph_df.join(region_df[region + "price25ma"])
```

## now as you can see, it's alot faster!

```python
graph_df.dropna().plot(figsize = (10,5), legend = False)
```
